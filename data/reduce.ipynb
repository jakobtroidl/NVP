```python
from PIL import Image
import os
import re
import glob
```

```python
in_path = "/home/jakobtroidl/Desktop/NVP/data/readysetgo/"
out_path = "/home/jakobtroidl/Desktop/NVP/data/readysetgo-25/"
```

```python
def sort_key(path):
    # Extract numbers and convert them to integers
    numbers = re.findall(r'\d+', path)
    return [int(num) for num in numbers]
```

```python
# loop over all files in in_path
video_path = os.path.join(in_path, "*.png")
files = sorted(glob.glob(video_path), key=sort_key)

n = 4

for file in files:
    # print("File:", file)
    file_number = int(re.search(r'(\d+)(?=\.\w+$)', file).group())

    if (file_number - 1) % n == 0:
        print("Saving file:", file_number)
        image = Image.open(file)
        image.save(os.path.join(out_path, os.path.basename(file)))



```
